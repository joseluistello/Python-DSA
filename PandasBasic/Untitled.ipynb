```python
# plot sin(x) and cos^2(x) from 0 to 2pi

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)**2

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos^2(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# create a regression plot with random numbers

import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y)
plt.show()
```

```python
# create a linear regression with a random set of numbers

import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
plt.show()
```
