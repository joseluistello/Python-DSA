```python
## Calling just one column / variable
df["NOCCode"]
```

```python
## Calling two columns / variables
df[["NOCCode", "Bronze Medal"]]
```

```python
## Subsetting multiple columns
cols_to_subset = ["NOCCode", "Bronze Medal"]
df[cols_to_subset]
```

```python
# Select the rows with values of 10 bronze medals or more
df[df["Bronze Medal"] > 10]
```

```python
# Subsetting based on text
df[df["NOCCode"] == "FRA"]
```

```python
# Multiple condition subsetting
country = df["Team/NOC"] == "Japan"
idd = df["NOCCode"] == "JPN"
df[country & idd]
```

```python
# Subsetting between some values
between_10 = df["Gold Medal"].isin(["4", "10"])
df[between_10]
```

```python
## Transform column
df["Gold Medal"] = df["Gold Medal"] / 2 
df
```

```python
## Multiple Manipulation
MoreBronzeMedals = df[df["Bronze Medal"] > 10]
MoreBronze = MoreBronzeMedals.sort_values("Bronze Medal", ascending = False)
MoreBronze[["Team/NOC", "NOCCode", "Bronze Medal"]]
```
